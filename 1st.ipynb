```python
x = torch.tensor(6.7) #input feature
y = torch.tensor(0.0) #True label (binary)

w = torch.tensor(1.0) #weight
b = torch.tensor(0.0) #bias
```

```python
#Binary Cross-Entropy Loss for Scaler
def bce_loss(predictions, targets):
    epilon = 1e-7  # to avoid log(0)
    predictions = torch.clamp(predictions, epilon, 1 - epilon)
    return - (targets * torch.log(predictions) + (1 - targets) * torch.log(1 - predictions))
```

```python
z = w * x + b  #weighted sum
y_pred = torch.sigmoid(z)  #predicted probability

#Compute Binary Cross-Entropy Loss
loss = bce_loss(y_pred, y)
```

```python
loss
```

```python
#Derivatives:
# 1. dl/d(y_pred) loss with respect to prediction (y_pred)
dloss_dy_pred = - (y / y_pred) + (1 - y) / (1 - y_pred)

# 2. dy_pred/dz prediction (y_pred) with respect to sum z (sigmoid derivative)
dy_pred_dz = y_pred * (1 - y_pred)

#` 3. dz/dw sum z with respect to w and b
dz_dw = x #dz/db = x
dz_db = 1 #dz/db = 1 (bias contributes directly to z)

dl_dw = dloss_dy_pred * dy_pred_dz * dz_dw
dl_db = dloss_dy_pred * dy_pred_dz * dz_db
```

```python
print("Derivative of loss with respect to weight w:", dl_dw)
print("Derivative of loss with respect to bias b:", dl_db)
```
